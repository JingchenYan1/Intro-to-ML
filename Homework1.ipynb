```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Use the URL for the raw CSV data
url = 'https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/main/Dataset/D3.csv'
df = pd.read_csv(url)

X1 = df.values[:, 0]
X2 = df.values[:, 1]
X3 = df.values[:, 2]
y = df.values[:, 3]

m = len(y)

X_0 = np.ones((m, 1))
X_1 = X1.reshape(m, 1)
X_1 = np.hstack((X_0, X_1))

X_2 = X2.reshape(m, 1)
X_2 = np.hstack((X_0, X_2))

X_3 = X3.reshape(m, 1)
X_3 = np.hstack((X_0, X_3))

def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    m = len(y)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J

theta1 = np.zeros(2)
theta2 = np.zeros(2)
theta3 = np.zeros(2)


def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

iterations = 1500
alpha = 0.01

theta1, cost_history1 = gradient_descent(X_1, y, theta1, alpha, iterations)
theta2, cost_history2 = gradient_descent(X_2, y, theta2, alpha, iterations)
theta3, cost_history3 = gradient_descent(X_3, y, theta3, alpha, iterations)

# X1
print(f"Linear Model for X1: y = {theta1[1]:.4f} * X1 + {theta1[0]:.4f}")
print(f"Final loss for X1 model: {cost_history1[-1]:.4f}")
# model
plt.scatter(X_1[:, 1], y, color='red', marker='+', label='X1')
plt.plot(X_1[:, 1], X_1.dot(theta1), color='green', label='Linear Model for X1')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Model for X1')
plt.legend()
plt.show()
# loss
plt.plot(range(1, iterations + 1), cost_history1, color='blue', label='Loss for X1 Model')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X1 model')
plt.show()

# X2
print(f"Linear Model for X2: y = {theta2[1]:.4f} * X2 + {theta2[0]:.4f}")
print(f"Final loss for X2 model: {cost_history2[-1]:.4f}")
# model
plt.scatter(X_2[:, 1], y, color='red', marker='+', label='X2')
plt.plot(X_2[:, 1], X_2.dot(theta2), color='green', label='Linear Model for X2')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('Linear Model for X2')
plt.legend()
plt.show()
# loss
plt.plot(range(1, iterations + 1), cost_history2, color='blue', label='Loss for X2 Model')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X2 model')
plt.show()

# X3
print(f"Linear Model for X3: y = {theta3[1]:.4f} * X3 + {theta3[0]:.4f}")
print(f"Final loss for X3 model: {cost_history3[-1]:.4f}")
# model
plt.scatter(X_3[:, 1], y, color='red', marker='+', label='X3')
plt.plot(X_3[:, 1], X_3.dot(theta3), color='green', label='Linear Model for X3')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('Linear Model for X3')
plt.legend()
plt.show()
# loss
plt.plot(range(1, iterations + 1), cost_history3, color='blue', label='Loss for X3 Model')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for X3 model')
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Use the URL for the raw CSV data
url = 'https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/main/Dataset/D3.csv'
df = pd.read_csv(url)

X1 = df.values[:, 0]
X2 = df.values[:, 1]
X3 = df.values[:, 2]
y = df.values[:, 3]

m = len(y)

X_0 = np.ones((m, 1))
X_1 = X1.reshape(m, 1)
X_2 = X2.reshape(m, 1)
X_3 = X3.reshape(m, 1)

X = np.hstack((X_0, X_1, X_2, X_3))

def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    m = len(y)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J

theta = np.zeros(4)

def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

iterations = 5000
alpha = 0.01

theta, cost_history = gradient_descent(X, y, theta, alpha, iterations)

print(f"Linear Model: y = {theta[3]:.4f} * X3 + {theta[2]:.4f} * X2 + {theta[1]:.4f} * X1 + {theta[0]:.4f}")
print(f"Final loss for linear model: {cost_history[-1]:.4f}")

# loss
plt.plot(range(1, iterations + 1), cost_history, color='blue', label='Loss for Linear Model')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent for Linear model')
plt.show()

# Nex X
new_X = np.array([[1, 1, 1], [2, 0, 4], [3, 2, 1]])
new_X = np.hstack((np.ones((new_X.shape[0], 1)), new_X))
new_y = new_X.dot(theta)

print(f"Predicted y for X = (1, 1, 1): {new_y[0]:.4f}")
print(f"Predicted y for X = (2, 0, 4): {new_y[1]:.4f}")
print(f"Predicted y for X = (3, 2, 1): {new_y[2]:.4f}")
```
